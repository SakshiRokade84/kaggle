VISUALISING THE DATASET


```python
sns.scatterplot(data= dataset, x = "TV", y="Sales", edgecolor = "black")
plt.show()
```

```python
x = dataset[["TV"]]
y = dataset[["Sales"]]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state=30)
```

```python
x_train.shape, x_test.shape,y_train.shape, y_test.shape
```

```python
print(x_train.shape, y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.predict([[8.7]])
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
a = np.array((lr.coef_*8.7)+lr.intercept_)
a
```

```python
y_predict = lr.predict(x)
```

```python
sns.scatterplot(data= dataset, x = "TV", y = "Sales", label = "original_data")
plt.plot(x, y_predict, label ="prediction_line", c = "red")
plt.legend()
plt.show()
```
